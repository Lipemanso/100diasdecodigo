```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({

    'exibicoes_painel': [5, 10, 3, 7, 12],
    'cliques_em_odds': [2, 6, 1, 3, 8],
    'tempo_na_plataforma': [30, 70, 20, 50, 90],
    'trocas_de_canal': [1, 4, 0, 2, 5]
})
```

```python
matriz_corr = df.corr(method='pearson')
print(matriz_corr)
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f' )
plt.title("Correlação")
plt.show()
```
